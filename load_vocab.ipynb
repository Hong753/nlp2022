```python
import pickle as pkl
```

```python
with open('./data/CUB_200_2011/birds/captions_DAMSM.pickle', 'rb') as f:
    captions = pkl.load(f)
```

```python
vocabs = captions[2]

# CUB_200_2011 vocab dictionary
for word in vocabs:
    print(vocabs[word])
```

```python
ban_list = ["and", "is", "bird", "this", "with", "a", "an", "the", "has", "looks", "that", "of", "are", "have", "on", \
            "as", "well", "its", "it", "s", "in", "also"]
```

```python
# add 
results = {}
for idx in range(len(captions[1])):
    results[idx] = {"attributes": [[]], "attributes_txt": [[]], "original_sentence": captions[1][idx]}
    for word in captions[1][idx]:
        if vocabs[word] not in ban_list:
            results[idx]["attributes"][-1].append(word)
            results[idx]["attributes_txt"][-1].append(vocabs[word])
        else:
            if len(results[idx]["attributes"][-1]) != 0:
                results[idx]["attributes"].append([])
                results[idx]["attributes_txt"].append([])
            
    sentence = ""
    for word in captions[1][idx]:
        sentence = sentence + vocabs[word]
        sentence += " "
    sentence = sentence.rstrip()
```

```python
import joblib
joblib.dump(results, "./data/CUB_200_2011/birds/prepreocessed_captions.pkl")
```
